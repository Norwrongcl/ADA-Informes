##2.2 Descripción del algoritmo
Tal como fue mencionado, para crear una solución a este problema se utilizará un algoritmo tipo greedy, el cual es inexacto pero al ser una solución general y no enfocada a un solo caso, intenta resolver la mayoría de casos. 
### Implementación greedy
Este método recibe un **layout** que contiene los stacks a ordenar. Este **layout** se genera a partir de `S` que es la cantidad de stacks, `H` la cual es la altura máxima a la que pueden llegar los contenedores en cada stack y `N`que es la cantidad de contenedores que hay repartida por los stacks. Destacar que el algoritmo tambien recibe estas 3 últimas variables. En general, los pasos que sigue esta solución es la siguiente:

**Se define como TOP el número que se encuentra en la parte superior del stack**

1. Recibe los datos. El generador de instancias no debería entregar en teoría stacks ordenados, asi que no comprobamos antes de realizar permutaciones.

2. Se crea una variable `mintop1` que irá almacenando el índice del stack con el **menor TOP**. Este será el stack que vaciaremos primero para luego rellenarlo con "contenedores" ya ordenados

3. Luego, se agrega este **índice** a una lista `expres` para que este stack no sea considerado en nuevas búsquedas de **TOPS**, así nos aseguramos de que lo que está resuelto **NO** se toca.

4. Se crea una variable `maxtop1` que irá almacenando el índice del stack con el **mayor TOP**, esto se hace con el fin de intentar mandar este "contenedor" al fondo del stack vacíado.

5. Se crea una variable `mintop2` esta se encargara de almacenar índice del stack con el siguiente **menor TOP** y se valida que este stack no haya alcanzado la altura máxima aún. La idea de buscar el siguiente TOP menor, es para que cuando vaciemos nuestro `lay.stacks[mintop1]`, ir dejando los contenedores más grandes en la parte superior de los stacks para luego dejarlos ordenados en el stack vacío.

6. Luego se manda el **TOP** de `lay.stacks[maxtop1]` al stack vacío. 

7. Se continúa buscando otro `maxtop1` para colocar en nuestro stack seleccionado hasta que o bien se llene o no queden más opciones de movimientos, esto último ocurre cuando ya no quedan más **TOPS** menores al **TOP** de nuestro stack.

8. Si el **Layout** aún no se encuentra ordenado, repetimos el procedimiento.
